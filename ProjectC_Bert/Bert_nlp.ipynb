```python
from transformers import *
from snownlp import SnowNLP
import numpy
# Load model, model config and tokenizer via Transformers
custom_config = AutoConfig.from_pretrained("distilbert-base-multilingual-cased")
custom_config.output_hidden_states=True
custom_tokenizer = AutoTokenizer.from_pretrained("distilbert-base-multilingual-cased")
custom_model = AutoModelWithLMHead.from_pretrained("distilbert-base-multilingual-cased", config=custom_config)

from Summarizer_ProjectC import Summarizer
from Summarizer_ProjectC.coreference_handler import CoreferenceHandler
model = Summarizer(custom_model=custom_model, custom_tokenizer=custom_tokenizer)
```

```python
def do_Summarizer(title,body):
    print("\n摘要分析 -\n")
    result=model(body)
    full=u''.join(result)
    print("標題:"+title+"\n本文:\n"+body+"\n摘要:\n"+full+"\n本文字數:"+str(len(body))+"\n摘要字數:"+str(len(full)))
    return str(full)
```

```python
#情緒分析
def emotion_Analysis(title,body):
    print("\n情緒分析 -\n")
    s =[SnowNLP(title),SnowNLP(body)]
    total_Emotion=0
    type_Emotion=0
    for s_type in s:
        toal_Emotion=0
        for sent in s_type.sentences:
            emotion = SnowNLP(sent)
            total_Emotion += emotion.sentiments
        print("句子數量:"+str(len(s_type.sentences)))
        total_Emotion/=len(s_type.sentences)
        print("平均情緒指數:"+str(total_Emotion))
        for sent in s_type.sentences:
            emotion = SnowNLP(sent)
            print(sent)
            print(emotion.sentiments)
        type_Emotion+=total_Emotion
    type_Emotion/=3
    print("標題與內文情緒指數:"+str(type_Emotion))
    return total_Emotion
```
