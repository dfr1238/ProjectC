```python
sentence_list = [articles]
word_sentence_list = ws(sentence_list,
            sentence_segmentation=True,
            segment_delimiter_set={'?', '？', '!', '！', '。', ',','，', ';', ':', '、'})
pos_sentence_list = pos(word_sentence_list)
entity_sentence_list = ner(word_sentence_list, pos_sentence_list)
```

```python
def print_word_pos_sentence(word_sentence, pos_sentence):
    assert len(word_sentence) == len(pos_sentence)
    for word, pos in zip(word_sentence, pos_sentence):
        print(f"{word}({pos})", end="\u3000")
    print()
    return
    
for i, sentence in enumerate(sentence_list):
    print()
    print(f"'{sentence}'")
    print_word_pos_sentence(word_sentence_list[i],  pos_sentence_list[i])
    for entity in sorted(entity_sentence_list[i]):
        print(entity)
```

```python
type(word_sentence_list)
word_sentence_list
```

```python
word_freq = {}
punctuation='（?？!！。,，;:、）「」/：'
for word in word_sentence_list[0]:
    if word not in punctuation:
        if word not in word_freq.keys():
            word_freq[word] = 1
        else:
            word_freq[word] +=1
```

```python
word_freq
```

```python
max_freq= max(word_freq.values())
print(max_freq)
```

```python
for word in word_freq.keys():
    word_freq[word]=word_freq[word]/max_freq
```

```python
print(word_freq)
```

```python
articles_sentences=[]
temp_str=""
for letter in articles:
    if(letter == "。"):
        temp_str+=letter
        articles_sentences.append(temp_str)
        temp_str=""
    else:
        temp_str+=letter
```

```python
articles_sentences
```

```python
articles_sentences_weight = {}
#print(len(articles_sentences))
for i in range(0, len(articles_sentences)):
    temp_list=[articles_sentences[i]]
    #print(temp_list)
    sentence = ws(temp_list,
            sentence_segmentation=True,
            segment_delimiter_set={'?', '？', '!', '！', '。', ',','，', ';', ':', '、'})
    print(sentence)
    for word in sentence[0]:
        if word in word_freq.keys():
            if articles_sentences[i] not in articles_sentences_weight.keys():
                articles_sentences_weight[articles_sentences[i]] = word_freq[word]
            else:
                articles_sentences_weight[articles_sentences[i]] += word_freq[word]
```

```python
articles_sentences_weight
```

```python
from heapq import nlargest
```

```python
select_length = int(len(articles_sentences)*0.3)
select_length
```

```python
summary = nlargest(select_length, articles_sentences_weight, key = articles_sentences_weight.get)
```

```python
final_summary=[word for word in summary]
summary = ' '.join(final_summary)
```
